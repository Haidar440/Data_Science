# Data Visualization with Seaborn, Matplotlib, and Plotly
This notebook demonstrates how to use Seaborn, Matplotlib, and Plotly for data visualization and analysis. Each code cell is preceded by a markdown cell explaining its purpose and usage with examples.

## Importing Seaborn
Seaborn is a powerful Python library for statistical data visualization built on top of matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.
**Example:**
```python
import seaborn as sns
```

````python
df = sns.load_dataset('iris')
df
````

## Loading the Iris Dataset
Seaborn provides built-in datasets for practice and demonstration. Here, we load the famous Iris dataset, which contains measurements of iris flowers from three different species.
**Example:**
```python
df = sns.load_dataset('iris')
df.head()  # Display the first few rows
```

````python
sns.scatterplot(df,x='sepal_length',y='petal_length',hue='species')
````

## Scatter Plot with Seaborn
A scatter plot visualizes the relationship between two numerical variables. The `hue` parameter colors the points by species, making it easy to distinguish between different classes.
**Example:**
```python
sns.scatterplot(df, x='sepal_length', y='petal_length', hue='species')
```
- Each point represents a flower sample.
- Colors indicate the species.

````python
import matplotlib.pyplot as plt
````

## Importing Matplotlib
Matplotlib is a foundational plotting library in Python. Seaborn is built on top of matplotlib and often requires it for advanced customization.
**Example:**
```python
import matplotlib.pyplot as plt
```

````python
plt.plot(df.sepal_length,df.petal_length,'o')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
````

## Scatter Plot with Matplotlib
This code creates a scatter plot using matplotlib to visualize the relationship between sepal length and petal length.
**Example:**
```python
plt.plot(df.sepal_length, df.petal_length, 'o')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```
- `'o'` specifies circle markers for each data point.

````python
plt.hist(df['sepal_length'],alpha=0.5)
````

## Histogram of Sepal Length
A histogram shows the distribution of a numerical variable by dividing the data into bins and displaying the frequency of values in each bin.
**Example:**
```python
plt.hist(df['sepal_length'], alpha=0.5)
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```
- `alpha=0.5` sets the transparency of the bars.

````python
sns.distplot(df['sepal_length'])
````

## Distribution Plot with Seaborn
A distribution plot (distplot) shows the distribution of a numerical variable as a combination of a histogram and a kernel density estimate (KDE).
**Example:**
```python
sns.distplot(df['sepal_length'])
```
- Shows both the histogram and a smooth curve representing the data distribution.
- Note: In recent versions of seaborn, use `sns.histplot` or `sns.kdeplot` instead.
