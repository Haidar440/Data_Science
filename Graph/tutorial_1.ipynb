The following code plots a histogram for all numerical columns in the DataFrame:

## Plotting a Histogram for Sepal Length
A **histogram** is a graphical representation that shows the distribution of a numerical variable by dividing the data into bins and displaying the frequency of values in each bin. It helps you understand how your data is spread out and whether it is skewed, centered, or has outliers.

The following code plots a histogram for the `sepal_length` column in the Iris dataset:

## Visualizing Distributions with DataFrame.hist()
The `df.hist()` function creates histograms for each numerical column in the DataFrame. This helps you visualize the distribution of each variable at once.

**Parameters used:**
- `figsize=(15,8)`: Sets the size of the entire figure.
- `color='g'`: Sets the color of the bars to green.
- `bins=20`: Divides the data into 20 intervals (bins).
- `alpha=1`: Sets the opacity of the bars (1 means fully opaque).

**Example:**
```python
df.hist(figsize=(15,8), color='g', bins=20, alpha=1)
```
This will show a grid of histograms, one for each numeric column in your DataFrame.

## Visualizing Data Distribution with a Box Plot
A **box plot** (or box-and-whisker plot) is a graphical representation that shows the distribution, spread, and outliers of numerical data. It displays the median, quartiles, and possible outliers for each column.

The following code creates a box plot for all numerical columns in the DataFrame:
```python
df.plot(kind='box', figsize=(15,8), title='Box Plot of Iris Dataset')
```
- The box shows the interquartile range (middle 50% of the data).
- The line inside the box is the median (middle value).
- Whiskers show the range of the data, except for outliers.
- Dots outside the whiskers represent outliers.

Box plots are useful for comparing distributions and spotting outliers in your data.

## Descriptive Statistics with DataFrame.describe()
The `df.describe()` function generates summary statistics for each numerical column in the DataFrame. It provides useful information such as:
- **count**: Number of non-null values
- **mean**: Average value
- **std**: Standard deviation (spread of the data)
- **min**: Minimum value
- **25%**: First quartile (25% of data below this value)
- **50%**: Median (middle value)
- **75%**: Third quartile (75% of data below this value)
- **max**: Maximum value

This helps you quickly understand the central tendency, spread, and shape of your data.

## Visualizing Data with an Area Plot
An **area plot** displays quantitative data graphically, showing the magnitude of values over a continuous interval or time period. It is similar to a line plot, but the area between the line and the axis is filled with color, making it easy to compare multiple variables.

The following code creates an area plot for all numeric columns in the DataFrame:
```python
df.plot(kind='area', figsize=(15,8), title='Area Plot of Iris Dataset')
```
- Each colored area represents a column's values over the index (e.g., row number or date).
- Useful for visualizing cumulative totals or comparing trends between columns.

````python
df.plot(kind = 'area' , figsize = (20,10),stacked = False)
````

## Creating a Non-Stacked Area Plot
By default, area plots are stacked, meaning each column's area is drawn on top of the previous one. Setting `stacked=False` creates a non-stacked area plot, where each column's area is independent, making it easier to compare trends.

**Example:**
```python
df.plot(kind='area', figsize=(20,10), stacked=False)
```
- Each colored area represents a column's values, not stacked on top of each other.

## Visualizing Density with a Hexbin Plot
A **hexbin plot** is used to visualize the density of points in a scatter plot, especially for large datasets. It divides the plot area into hexagons and colors them based on the number of points in each hexagon.

The following code creates a hexbin plot of `sepal_length` vs. `petal_length`, with color intensity based on `sepal_width`:
```python
df.plot.hexbin(x='sepal_length', y='petal_length', gridsize=10, cmap='Greens', C='sepal_width')
```
- `gridsize=10`: Sets the number of hexagons in the x-direction.
- `cmap='Greens'`: Sets the color map.
- `C='sepal_width'`: Colors hexagons by the average sepal width in each bin.
- Useful for visualizing point density and relationships in large datasets.

## Interactive Plotting with Cufflinks and Plotly
The `df.iplot()` function from the cufflinks library creates interactive plots using Plotly. It works best with default arguments. If you want to customize colors, use valid CSS color strings (e.g., `'rgb(255,153,51)'` or `'#ff9933'`).
**Note:** Avoid passing numpy or pandas objects as color arguments. Only use valid color strings as described in the [Plotly documentation](https://plotly.com/python/css-colors/).


## Creating Interactive Line Plots with Plotly Express
Plotly Express (`px`) is a high-level interface for creating interactive plots. You can use `px.line()` to plot one or more numeric columns from your DataFrame.
**Example:**
```python
import plotly.express as px
numeric_cols = df.select_dtypes(include='number').columns
fig = px.line(df, y=numeric_cols, title="Line Plot with Plotly Express")
fig.show()
```
- This will create an interactive line plot for all numeric columns in your DataFrame.

````python
fig2 = px.scatter(df,x='sepal_length', y='petal_length', title="Iris Dataset Scatter Plot", )
fig2.show()
````

````python
# Plotly Express does not have a 'bubble' function. Use px.scatter with 'size' for bubble plots.
fig2 = px.scatter(
	df,
	x='sepal_length',
	y='petal_length',
	size='sepal_width',  # Bubble size
	title="Iris Dataset Bubble Plot"
)
fig2.show()
````

````python
fig2 = px.scatter_3d(
	df,
	x='sepal_length',
	y='sepal_width',
    z='petal_length',
	size='sepal_width',  # Bubble size
	title="Iris Dataset Bubble Plot"
)
fig2.show()
````

````python
import plotly.express as px
fig2 = px.scatter_3d(
    df.head(),
    x='sepal_length',
    y='sepal_width',
    z='petal_length',
    size='sepal_width',  # Bubble size
    title="Iris Dataset Bubble Plot",
    color_discrete_sequence=['orange']  # Set all points to orange
)
fig2.show()
````
