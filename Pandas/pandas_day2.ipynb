```python
df2 = pd.DataFrame(data,columns=['id','repository_url',
                           'labels_url','user','user[login]','user[node_id]','user[id]'])
```

```python
# Use a different filename to avoid permission issues
import datetime
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
filename = f'json_data_{timestamp}.csv'

try:
    df2.to_csv(filename, index=False)
    print(f"Successfully saved data to {filename}")
except PermissionError as e:
    print(f"Permission error: {e}")
    print("Make sure the file is not open in Excel or another application")
except Exception as e:
    print(f"Error saving file: {e}")



pd.DataFrame(data)
```
